### Mostra a quantidade de nulos em todas as colunas.

```python
for col in df_diretor.columns:
    print(col, df_diretor[col].dtype, "nulos:", df_diretor[col].isnull().sum())
```

### Mostra a porcentagem de nulos em cada coluna.


```python
for col in df_diretor.columns:
    nulos = df_diretor[col].isnull().sum()
    total = len(df_diretor)
    perc = nulos / total * 100
    if perc > 10:
        dtype_str = str(df_diretor[col].dtype)  # converte o dtype para string
        print(f"{col:30} {dtype_str:10} nulos: {nulos:4} ({perc:5.1f}%)")
```
